We use the basic transformer and default transformer hyperparameters to begin with. Change this to change the model

```python
MODEL = 'transformer'
HPARAMS = 'transformer_base'
PROBLEM = 'mimic_discharge_summaries'

TRAIN_DIR = '../data/t2t/output'
DATA_DIR = '../data/t2t/data'
```

```python
# Init Hparams object from T2T Problem
hparams = create_hparams(HPARAMS)

for item in ast.literal_eval(str(hparams)):
    print(item[0], ":", item[1])
```

```python
# # Make Changes to Hparams if necessary

# hparams.batch_size = 1024
# hparams.learning_rate_warmup_steps = 45000
# hparams.learning_rate = .4
```

```python
RUN_CONFIG = create_run_config(
    model_name = MODEL,
    model_dir = TRAIN_DIR, # Location of where model file will be stored
    num_gpus = 4
)

# Create Tensorflow Experiment Object
tensorflow_exp_fn = create_experiment(
    run_config=RUN_CONFIG,
    hparams=hparams,
    model_name=MODEL,
    problem_name=PROBLEM,
    data_dir=DATA_DIR, 
    train_steps=400000, # Total number of train steps for all Epochs
    eval_steps=100 # Number of steps to perform for each evaluation
)
```

```python
# Kick off Training
tensorflow_exp_fn.train_and_evaluate()
```
